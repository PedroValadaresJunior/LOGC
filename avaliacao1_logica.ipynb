Proposta: Desenvolver um algoritmo em Python que gere a tabela-verdade que demostre a
propriedade do

    `(𝑃 → 𝑄) ∧ (𝑅 → 𝑆) ∧ (¬𝑄 ∨ ¬𝑆) → (¬𝑃 ∨ ¬𝑅)`

Enviar o link do Gut - Hub ou o arquivo em Python


```python
# (𝑃 → 𝑄) ∧ (𝑅 → 𝑆) ∧ (¬𝑄 ∨ ¬𝑆) → (¬𝑃 ∨ ¬𝑅)

print(f' P | Q | R | S | ¬𝑃 | ¬Q | ¬R | ¬S | (𝑃 → 𝑄) | (𝑅 → 𝑆) | (¬𝑄 ∨ ¬𝑆) | (¬𝑃 ∨ ¬𝑅) | (𝑃 → 𝑄) ∧ (𝑅 → 𝑆) ∧ (¬𝑄 ∨ ¬𝑆) → (¬𝑃 ∨ ¬𝑅)')

n = 0
proposicao = ['P', 'Q', 'R', 'S']
target = 2 ** len(proposicao)
lista = ['V', 'F']

for a in range(2):

    for b in range(2):

        for c in range(2):

            for d in range(2):

                P = lista[a]
                Q = lista[b]
                R = lista[c]
                S = lista[d]

                # ¬𝑃
                NP = 'F' if P == 'V' else 'V'

                # ¬Q
                NQ = 'F' if Q == 'V' else 'V'

                # ¬R
                NR = 'F' if R == 'V' else 'V'

                 # ¬S
                NS = 'F' if S == 'V' else 'V'

                # (𝑃 → 𝑄)
                PQ = 'F' if (P == 'V' and Q == 'F') else 'V'

                # (𝑅 → 𝑆)
                RS = 'F' if (R == 'V' and S == 'F') else 'V'

                # (¬𝑄 ∨ ¬𝑆)
                NQNS = 'V' if NQ == 'V' or NS == 'V' else 'F'

                # (¬𝑃 ∨ ¬𝑅)
                NPNR = 'V' if NP == 'V' or NR == 'V' else 'F'

                # (𝑃 → 𝑄) ∧ (𝑅 → 𝑆)
                PQRS = 'V' if PQ == 'V' and RS == 'V' else 'F'

                # (𝑃 → 𝑄) ∧ (𝑅 → 𝑆) ∧ (¬𝑄 ∨ ¬𝑆)
                PQRSNQNS = 'V' if PQRS == 'V' and NQNS == 'V' else 'F'

                # (𝑃 → 𝑄) ∧ (𝑅 → 𝑆) ∧ (¬𝑄 ∨ ¬𝑆) → (¬𝑃 ∨ ¬𝑅)
                PQRSNQNSNPNR = 'F' if PQRSNQNS == 'V' and NPNR == 'F' else 'V'
                n+=1 if PQRSNQNSNPNR == 'V' else 0

                print(f' {P} | {Q} | {R} | {S} |  {N𝑃} |  {NQ} |  {NR} |  {NS} |    {PQ}    |     {RS}   |      {NQNS}    |    {NPNR}      |        {PQRSNQNSNPNR} ')
   


if n == 0:
    print("\nA tabela-verdade é uma contradição")

elif n == target:
    print("\nA tabela-verdade é uma tautologia")

else:
    print("\nA tabela-verdade é uma contingência")
```
